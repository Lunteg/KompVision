```python
import tkinter as tk
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import (
    FigureCanvasTkAgg, NavigationToolbar2Tk)
from matplotlib.figure import Figure
import numpy as np
import cv2

def change_saturation(img_name, k):
    img = cv2.imread(img_name)
    img_LAB = cv2.cvtColor(img, cv2.COLOR_BGR2LAB) # от зеленого до красного, вторая — от синего до желтого
    # img_LAB = np.array([[[L00, a00, b00], [L01, a01, b01]], [[L10, a10, b10], [L11, a11, b11]]])

    n = len(img_LAB)
    m = len(img_LAB[0])
    l = []
    a = []
    b = []
    for i in range(n):
        for j in range(m): 
            l.append(img_LAB[i][j][0])
            a.append(min(255, img_LAB[i][j][1] * k))
            b.append(min(255, img_LAB[i][j][2] * k))
    
    l = np.array(l)
    a = np.array(a)
    b = np.array(b)

    a = a.astype(np.uint8)
    b = b.astype(np.uint8)

    a = np.reshape(a, (n, m))
    b = np.reshape(b, (n, m))
    l = np.reshape(l, (n, m))

    img = cv2.merge([l, a, b])
    return cv2.cvtColor(img, cv2.COLOR_LAB2BGR)

def change_brightness(img_name, k):
    img = cv2.imread(img_name)
    img_LAB = cv2.cvtColor(img, cv2.COLOR_BGR2LAB) 

    n = len(img_LAB)
    m = len(img_LAB[0])
    l = []
    a = []
    b = []

    for i in range(n):
        for j in range(m): 
            l.append(min(255, img_LAB[i][j][0] + k))
            a.append(img_LAB[i][j][1])
            b.append(img_LAB[i][j][2])
    l = np.array(l)
    a = np.array(a)
    b = np.array(b)

    l = l.astype(np.uint8)

    a = np.reshape(a, (n, m))
    b = np.reshape(b, (n, m))
    l = np.reshape(l, (n, m))

    img = cv2.merge([l, a, b])
    return cv2.cvtColor(img, cv2.COLOR_LAB2BGR)

class BaseWindow(tk.Tk):
    def __init__(self, img1, str1, img2, str2):
        super().__init__()
        self.geometry("750x750")
        self.title("Использование библиотеки TKinter - построение графиков")
        canvas = tk.Canvas(self, width = 0, height = -10)
        canvas.pack()
        fig, ax = plt.subplots(1, 2, figsize=(10, 20))

        ax[0].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
        ax[0].set_title(str1)

        ax[1].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
        ax[1].set_title(str2) 
       
        canvas = FigureCanvasTkAgg(fig, master=self)
        canvas.get_tk_widget().pack()        

if __name__ == "__main__":
    img_name = "bruklinskij-most.jpg"
    img = cv2.imread(img_name)
    #saturated_img = change_saturation(img_name, 1.05)
    #BaseWindow(img, 'Original picture', saturated_img, 'Changed saturation').mainloop()

    brightened_img = change_brightness(img_name, 20)
    BaseWindow(img, 'Original picture', brightened_img, 'Changed brightness').mainloop()
```
